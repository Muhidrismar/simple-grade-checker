## Buatlah sebuah program yang:

   1. Meminta input dari pengguna:

       - Nama mahasiswa

       - NIM mahasiswa

       - Nilai ujian (dalam angka, 0â€“100)

   2. Cek dan tampilkan tipe data dari setiap input.

   3. Gunakan if-elif-else untuk menentukan kategori nilai, dengan kriteria sebagai berikut:
   
```
Nilai | Kategori
85 - 100 | A (Sangat Baik)
75 - 84 | B (Baik)
60 - 74 | C (Cukup)
40 - 59 | D (Kurang)
< 40 | E (Sangat Kurang)
```

```
==== Ekspektasi Output Jawaban : =====
Masukkan nama mahasiswa: Sarah
Masukkan NIM: 12345678
Masukkan nilai ujian (0-100): 81

Nama: Sarah (type: <class 'str'>)
NIM: 12345678 (type: <class 'str'>)
Nilai: 81 (type: <class 'int'>)

Hasil Evaluasi:
Mahasiswa: Sarah (NIM: 12345678)
Nilai Ujian: 81
Kategori Nilai: B (Baik)
```

````python
namaMahasiswa = input("Masukan nama mahasiswa: ")
nim = input("Masukan NIM: ")
while True:
    try:
        nilai = int(input("Masukan nilai ujian (0-100): "))
        if 0 <= nilai <= 100:
            break
        else:
            print(" Nilai harus antara 0-100! ")
    except ValueError:
        print(" Input harus berupa angka! ")


if nilai >= 85:
    kategori = "A (Sangat Baik)"
elif nilai >= 75:
    kategori = "B (Baik)"
elif nilai >= 60:
    kategori = "C (Cukup)"
elif nilai >= 40:
    kategori = "D (Kurang)"
else:
    kategori = "E (Sangat Kurang)"

print(f"\nNama: {namaMahasiswa} (type: {str(type(namaMahasiswa))})")
print(f"NIM: {nim} (type: {str(type(nim))})")
print(f"Nilai: {nilai} (type: {str(type(nilai))})")


print(f"""
Hasil Evaluasi:
Mahasiswa: {namaMahasiswa} (NIM: {nim})
Nilai Ujian: {nilai}
Kategori Nilai: {kategori}
""")
````
